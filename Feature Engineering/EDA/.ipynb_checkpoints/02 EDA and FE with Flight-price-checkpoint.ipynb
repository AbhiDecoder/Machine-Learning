```python
df['duration']=df['duration'].astype(str)
```

```python
df['duration_hour']=df['duration'].str.split('.').str[0]
df['duration_minute']=df['duration'].str.split('.').str[1]
```

```python
df.head()
```

```python
df.info()
```

```python
df['duration_hour']=df['duration_hour'].astype(int)
df['duration_minute']=df['duration_minute'].astype(int)
```

```python
df.drop('duration',axis=1,inplace=True)
```

### OneHotEncoder for [airline , source_city , departure_time  , destination_city , stops , class]

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder()
```

```python
encoder.fit_transform(df[['airline' , 'source_city' , 'departure_time' , 'destination_city' , 'stops' , 'class']]).toarray()
```

```python
pd.DataFrame(encoder.fit_transform(df[['airline' , 'source_city' , 'departure_time' , 'destination_city' , 'stops' , 'class']]).toarray(),columns=encoder.get_feature_names_out())
```
