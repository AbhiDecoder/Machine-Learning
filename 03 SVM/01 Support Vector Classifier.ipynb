```python
# import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
# Lets Create Synthetic Data point
from sklearn.datasets import make_classification
'''Generate a random n-class classification problem'''
```

```python
X,y=make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,n_classes=2)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

```python
# Train test split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
#Implement Support Vector Classifier
from sklearn.svm import SVC
```

```python
svc=SVC() # kernal='rbf'
```

```python
# Fitting the model
svc.fit(X_train,y_train)
```

```python
# Prediction
y_pred=svc.predict(X_test)
```

```python
y_pred
```

```python
# Accuracy
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
# Using different kernals
svr=SVC(kernel='linear')
# Fitting the model
svc.fit(X_train,y_train)
# Prediction
y_pred=svc.predict(X_test)
# Accuracy
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
# Using different kernals
svr=SVC(kernel='poly')
# Fitting the model
svc.fit(X_train,y_train)
# Prediction
y_pred=svc.predict(X_test)
# Accuracy
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
# Using different kernals
svr=SVC(kernel='sigmoid')
# Fitting the model
svc.fit(X_train,y_train)
# Prediction
y_pred=svc.predict(X_test)
# Accuracy
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
# Hyper-parameter Tuning
from sklearn.model_selection import GridSearchCV
```

```python
#Defining Parameters
param_grid={'C':[0.1,1,10,100,1000],
            'gamma': [1,0.1,0.01,0.001,0.0001],
            'kernel': ['rbf','linear','poly','sigmoid']
    }
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
# Parameters selected
grid.best_params_
```

```python
# Prediction
y_pred4=grid.predict(X_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
```
